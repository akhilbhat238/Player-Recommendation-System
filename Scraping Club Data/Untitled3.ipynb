```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get("https://www.espn.in/football/table/_/league/TUR.1/season/2017")
soup = BeautifulSoup(page.content,'html.parser')
```

```python
tables = soup.findAll("table")
table_rows = tables[0].find_all('tr')
len(table_rows)
```

```python
outer_left_list=[]
for i in range(2,int(len(table_rows)/2)):
    inner_left_list=[]
    inner_left_list.append(table_rows[i].find('span').contents[0])
    inner_left_list.append(table_rows[i].find_all('a')[2].contents[0])
    outer_left_list.append(inner_left_list)
outer_left_list
left_df = pd.DataFrame(outer_left_list, columns =['Ranking','Team']) 
```

```python
outer_right_list = []
for i in range(22,len(table_rows)):    
    x = table_rows[i].find_all("span")
    inner_right_list=[]
    for j in range(0,len(x)):
        inner_right_list.append(x[j].contents[0])
    outer_right_list.append(inner_right_list)
right_df = pd.DataFrame(outer_right_list, columns =['GP','W','D','L','F','A','GD','P']) 
```

```python
complete_df = pd.concat([left_df, right_df], axis = 1)
```
