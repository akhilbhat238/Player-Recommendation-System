```python
import numpy as np
import pandas as pd
import re
import unicodedata
```

```python
bundesliga_valuations_df = pd.read_csv('..\\..\\Consolidated Data\\Bundesliga Club Valuations.csv',usecols=['Club','Total Market Value','League'])
bundesliga_valuations_df = bundesliga_valuations_df.rename(columns={'Club':'Team'})
```

```python
bundesliga_postions_df = pd.read_csv('..\\..\\Consolidated Data\\Bundesliga Average Positions.csv')
```

```python
bundesliga_combined_df = bundesliga_postions_df.merge(bundesliga_valuations_df,how='outer',on='Team')
```

```python
bundesliga_temp_df = bundesliga_combined_df.dropna()

bundesliga_next_df = pd.concat([bundesliga_combined_df,bundesliga_temp_df]).drop_duplicates(keep=False)
bundesliga_next_df = bundesliga_next_df.reset_index(drop=True)
```

```python
for i in range(bundesliga_next_df[bundesliga_next_df['Total Market Value'].isnull()].head(1).index[0],bundesliga_next_df[bundesliga_next_df['Total Market Value'].isnull()].tail(1).index[0]):
    for j in range(bundesliga_next_df[bundesliga_next_df['Avg_Pos'].isnull()].head(1).index[0],bundesliga_next_df[bundesliga_next_df['Avg_Pos'].isnull()].tail(1).index[0]):
        search = re.search(bundesliga_next_df.loc[i]['Team'],bundesliga_next_df.loc[j]['Team'],re.UNICODE)
        if search is not None:
            clean = search.group()
            bundesliga_next_df.loc[i] = bundesliga_next_df.loc[i].combine_first(bundesliga_next_df.loc[j])
            break
```

```python
bundesliga_next_df.loc[1] = bundesliga_next_df.loc[1].combine_first(bundesliga_next_df.loc[14])
bundesliga_next_df.loc[2] = bundesliga_next_df.loc[2].combine_first(bundesliga_next_df.loc[23])
bundesliga_next_df.loc[3] = bundesliga_next_df.loc[3].combine_first(bundesliga_next_df.loc[24])
bundesliga_next_df.loc[5] = bundesliga_next_df.loc[5].combine_first(bundesliga_next_df.loc[17])
bundesliga_next_df.loc[10] = bundesliga_next_df.loc[10].combine_first(bundesliga_next_df.loc[25])
bundesliga_next_df.loc[12] = bundesliga_next_df.loc[12].combine_first(bundesliga_next_df.loc[16])
```

```python
bundesliga_combined_df = pd.concat([bundesliga_combined_df.dropna(), bundesliga_next_df.dropna()])
```

```python
bundesliga_combined_df=bundesliga_combined_df.reset_index(drop=True)
```

```python
champ_league_data_df = pd.read_csv('..\\..\\Consolidated Data\\Champions_League_CLub_Level.csv')

champ_league_data_df.iloc[16, champ_league_data_df.columns.get_loc('Team')] = 'Schalke'

bundesliga_combined_df=bundesliga_combined_df.merge(champ_league_data_df,how='left',on='Team')

bundesliga_combined_df = bundesliga_combined_df.fillna(0)
```

```python
europa_league_data_df = pd.read_csv('..\\..\\Consolidated Data\\Europa_League_CLub_Level.csv')

europa_league_data_df.iloc[23, europa_league_data_df.columns.get_loc('Team')] = 'Schalke'

bundesliga_combined_df=bundesliga_combined_df.merge(europa_league_data_df,how='left',on='Team')

bundesliga_combined_df = bundesliga_combined_df.fillna(0)
```

```python
bundesliga_combined_df.columns = ['Team', 'Avg_Pos', 'Total Market Value', 'League', 'Range1_Win_CL',
       'Range2_Win_CL', 'Range3_Win_CL', 'Range4_Win_CL', 'Range5_Win_CL',
       'Range1_Run_CL', 'Range2_Run_CL', 'Range3_Run_CL', 'Range4_Run_CL',
       'Range5_Run_CL', 'Range1_SF_CL', 'Range2_SF_CL', 'Range3_SF_CL',
       'Range4_SF_CL', 'Range5_SF_CL', 'Range1_Win_EL', 'Range2_Win_EL',
       'Range3_Win_EL', 'Range4_Win_EL', 'Range5_Win_EL', 'Range1_Run_EL',
       'Range2_Run_EL', 'Range3_Run_EL', 'Range4_Run_EL', 'Range5_Run_EL',
       'Range1_SF_EL', 'Range2_SF_EL', 'Range3_SF_EL', 'Range4_SF_EL',
       'Range5_SF_EL']
```

```python
world_cup_data_df = pd.read_csv('..\\..\\Consolidated Data\\World Cup Data at Club Level.csv')

club_string = world_cup_data_df['Club']
for i in range(0,len(club_string)):
    club_string[i]=unicodedata.normalize('NFKD', club_string[i]).encode('ASCII', 'ignore').decode('utf-8')

world_cup_data_df['Club'] = club_string

world_cup_data_df.columns = ['Team','League','WC_Player_Count']

world_cup_data_df[world_cup_data_df['League'] ==  'Bundesliga']
```

```python
world_cup_data_df.iloc[0, world_cup_data_df.columns.get_loc('Team')] = 'FC Koln'
world_cup_data_df.iloc[1, world_cup_data_df.columns.get_loc('Team')] = 'Hoffenheim'
world_cup_data_df.iloc[2, world_cup_data_df.columns.get_loc('Team')] = 'Leverkusen'
world_cup_data_df.iloc[22, world_cup_data_df.columns.get_loc('Team')] = 'Hertha Berlin'
world_cup_data_df.iloc[30, world_cup_data_df.columns.get_loc('Team')] = 'Mainz'
world_cup_data_df.iloc[46, world_cup_data_df.columns.get_loc('Team')] = 'Schalke'
world_cup_data_df.iloc[54, world_cup_data_df.columns.get_loc('Team')] = 'Wolfsburg'
```

```python
bundesliga_temp_df=bundesliga_combined_df.merge(world_cup_data_df,how='outer',on='Team')

for i in range(bundesliga_temp_df[bundesliga_temp_df['Total Market Value'].isnull()].head(1).index[0],bundesliga_temp_df[bundesliga_temp_df['Total Market Value'].isnull()].tail(1).index[0]):
    for j in range(bundesliga_temp_df[bundesliga_temp_df['WC_Player_Count'].isnull()].head(1).index[0],bundesliga_temp_df[bundesliga_temp_df['WC_Player_Count'].isnull()].tail(1).index[0]):
        search = re.search(bundesliga_temp_df.loc[j]['Team'],bundesliga_temp_df.loc[i]['Team'])
        if search is not None:
            clean = search.group()
            bundesliga_temp_df.loc[i] = bundesliga_temp_df.loc[i].combine_first(bundesliga_temp_df.loc[j])
            break
```

```python
bundesliga_temp_df=bundesliga_temp_df.drop(['League_y'],axis=1)
bundesliga_temp_df['WC_Player_Count']=bundesliga_temp_df['WC_Player_Count'].fillna(0)
bundesliga_temp_df=bundesliga_temp_df.dropna()
bundesliga_combined_df = bundesliga_temp_df
```

```python
top_50_data_df = pd.read_csv('..\\..\\Consolidated Data\\Top 50 Players Club Level.csv')
top_50_data_df['Value in GBP'] = top_50_data_df['Value in GBP'].map(lambda x: x.lstrip('£').rstrip('m'))
x=top_50_data_df.groupby('Team').agg({'Player':'count'})
x=x.reset_index()
x.iloc[4, x.columns.get_loc('Team')] = 'Leverkusen'
bundesliga_temp_df=bundesliga_temp_df.merge(x,how='left',on='Team')
bundesliga_combined_df = bundesliga_temp_df.fillna(0)
```

```python
dfb_pokal_df = pd.read_csv('..\\..\\Consolidated Data\\DFB Pokal Records.csv')
dfb_winner = dfb_pokal_df[['Season','Winner']].groupby('Winner').count()
dfb_winner = dfb_winner.reset_index()
dfb_winner.columns = ['Team','DFB_Wins']
bundesliga_temp_df=bundesliga_combined_df
bundesliga_temp_df= pd.merge(bundesliga_temp_df,dfb_winner,how='left',on='Team')
bundesliga_temp_df=bundesliga_temp_df.fillna(0)
```

```python
dfb_runner = dfb_pokal_df[['Season','Runner']].groupby('Runner').count()
dfb_runner = dfb_runner.reset_index()
dfb_runner.columns = ['Team','DFB_Runners']
bundesliga_temp_df= pd.merge(bundesliga_temp_df,dfb_runner,how='left',on='Team')
bundesliga_temp_df=bundesliga_temp_df.fillna(0)
bundesliga_combined_df=bundesliga_temp_df
```
