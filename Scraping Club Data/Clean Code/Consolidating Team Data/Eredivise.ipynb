```python
import numpy as np
import pandas as pd
import re
import unicodedata
```

```python
eredivisie_valuations_df = pd.read_csv('..\\..\\Consolidated Data\\Eredivise Club Valuations.csv',usecols=['Club','Total Market Value','League'])
eredivisie_valuations_df = eredivisie_valuations_df.rename(columns={'Club':'Team'})
```

```python
eredivisie_postions_df = pd.read_csv('..\\..\\Consolidated Data\\Eredivise Average Positions.csv')
```

```python
eredivisie_combined_df = eredivisie_postions_df.merge(eredivisie_valuations_df,how='outer',on='Team')
```

```python
eredivisie_temp_df = eredivisie_combined_df.dropna()

eredivisie_next_df = pd.concat([eredivisie_combined_df,eredivisie_temp_df]).drop_duplicates(keep=False)
eredivisie_next_df = eredivisie_next_df.reset_index(drop=True)
```

```python
for i in range(eredivisie_next_df[eredivisie_next_df['Total Market Value'].isnull()].head(1).index[0],eredivisie_next_df[eredivisie_next_df['Total Market Value'].isnull()].tail(1).index[0]):
    for j in range(eredivisie_next_df[eredivisie_next_df['Avg_Pos'].isnull()].head(1).index[0],eredivisie_next_df[eredivisie_next_df['Avg_Pos'].isnull()].tail(1).index[0]):
        search = re.search(eredivisie_next_df.loc[i]['Team'],eredivisie_next_df.loc[j]['Team'],re.UNICODE)
        if search is not None:
            clean = search.group()
            eredivisie_next_df.loc[i] = eredivisie_next_df.loc[i].combine_first(eredivisie_next_df.loc[j])
            break
```

```python
eredivisie_next_df.loc[5] = eredivisie_next_df.loc[5].combine_first(eredivisie_next_df.loc[11])
```

```python
eredivisie_combined_df = pd.concat([eredivisie_combined_df.dropna(), eredivisie_next_df.dropna()])
```

```python
eredivisie_combined_df=eredivisie_combined_df.reset_index(drop=True)
```

```python
world_cup_data_df = pd.read_csv('..\\..\\Consolidated Data\\World Cup Data at Club Level.csv')

club_string = world_cup_data_df['Club']
for i in range(0,len(club_string)):
    club_string[i]=unicodedata.normalize('NFKD', club_string[i]).encode('ASCII', 'ignore').decode('utf-8')

world_cup_data_df['Club'] = club_string

world_cup_data_df.columns = ['Team','League','WC_Player_Count']

world_cup_data_df[world_cup_data_df['League'] ==  'Eredivisie']
```

```python
world_cup_data_df.iloc[2, world_cup_data_df.columns.get_loc('Team')] = 'AZ Alkmaar'
```

```python
eredivisie_temp_df=eredivisie_combined_df.merge(world_cup_data_df,how='outer',on='Team')

for i in range(eredivisie_temp_df[eredivisie_temp_df['Total Market Value'].isnull()].head(1).index[0],eredivisie_temp_df[eredivisie_temp_df['Total Market Value'].isnull()].tail(1).index[0]):
    for j in range(eredivisie_temp_df[eredivisie_temp_df['WC_Player_Count'].isnull()].head(1).index[0],eredivisie_temp_df[eredivisie_temp_df['WC_Player_Count'].isnull()].tail(1).index[0]):
        search = re.search(eredivisie_temp_df.loc[j]['Team'],eredivisie_temp_df.loc[i]['Team'])
        if search is not None:
            clean = search.group()
            eredivisie_temp_df.loc[i] = eredivisie_temp_df.loc[i].combine_first(eredivisie_temp_df.loc[j])
            break
```

```python
eredivisie_temp_df=eredivisie_temp_df.drop(['League_y'],axis=1)
eredivisie_temp_df['WC_Player_Count']=eredivisie_temp_df['WC_Player_Count'].fillna(0)
eredivisie_temp_df=eredivisie_temp_df.dropna()
eredivisie_combined_df = eredivisie_temp_df
```

```python
top_50_data_df = pd.read_csv('..\\..\\Consolidated Data\\Top 50 Players Club Level.csv')
top_50_data_df['Value in GBP'] = top_50_data_df['Value in GBP'].map(lambda x: x.lstrip('£').rstrip('m'))
x=top_50_data_df.groupby('Team').agg({'Player':'count'})
x=x.reset_index()
eredivisie_temp_df=eredivisie_temp_df.merge(x,how='left',on='Team')
eredivisie_combined_df = eredivisie_temp_df.fillna(0)
```

```python
knvb_cup_df = pd.read_csv('..\\..\\Consolidated Data\\KNVB Cup Records.csv')
knvb_winner = knvb_cup_df[['Season','Winner']].groupby('Winner').count()
knvb_winner = knvb_winner.reset_index()
knvb_winner.columns = ['Team','Copa_Wins']
eredivisie_temp_df=eredivisie_combined_df
eredivisie_temp_df= pd.merge(eredivisie_temp_df,knvb_winner,how='left',on='Team')
eredivisie_temp_df=eredivisie_temp_df.fillna(0)
```

```python
knvb_runner = knvb_cup_df[['Season','Runner']].groupby('Runner').count()
knvb_runner = knvb_runner.reset_index()
knvb_runner.columns = ['Team','Copa_Runners']
eredivisie_temp_df= pd.merge(eredivisie_temp_df,knvb_runner,how='left',on='Team')
eredivisie_temp_df=eredivisie_temp_df.fillna(0)
eredivisie_combined_df=eredivisie_temp_df
```
