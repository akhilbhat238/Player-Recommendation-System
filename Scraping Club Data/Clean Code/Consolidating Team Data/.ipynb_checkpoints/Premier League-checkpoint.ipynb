```python
import numpy as np
import pandas as pd
import re
import unicodedata
```

```python
prm_lg_valuations_df = pd.read_csv('..\\..\\Consolidated Data\\Premier League Club Valuations.csv',usecols=['Club','Total Market Value','League'])
prm_lg_valuations_df = prm_lg_valuations_df.rename(columns={'Club':'Team'})
```

```python
prm_lg_postions_df = pd.read_csv('..\\..\\Consolidated Data\\Premier League Average Positions.csv')
```

```python
prm_lg_combined_df = prm_lg_postions_df.merge(prm_lg_valuations_df,how='outer',on='Team')
```

```python
prm_lg_temp_df = prm_lg_combined_df.dropna()

prm_lg_next_df = pd.concat([prm_lg_combined_df,prm_lg_temp_df]).drop_duplicates(keep=False)
prm_lg_next_df = prm_lg_next_df.reset_index(drop=True)
```

```python
for i in range(prm_lg_next_df[prm_lg_next_df['Total Market Value'].isnull()].head(1).index[0],prm_lg_next_df[prm_lg_next_df['Total Market Value'].isnull()].tail(1).index[0]):
    for j in range(prm_lg_next_df[prm_lg_next_df['Avg_Pos'].isnull()].head(1).index[0],prm_lg_next_df[prm_lg_next_df['Avg_Pos'].isnull()].tail(1).index[0]):
        search = re.search(prm_lg_next_df.loc[i]['Team'],prm_lg_next_df.loc[j]['Team'],re.UNICODE)
        if search is not None:
            clean = search.group()
            prm_lg_next_df.loc[i] = prm_lg_next_df.loc[i].combine_first(prm_lg_next_df.loc[j])
            break
```

```python
prm_lg_next_df.loc[8] = prm_lg_next_df.loc[8].combine_first(prm_lg_next_df.loc[18])
prm_lg_next_df.loc[9] = prm_lg_next_df.loc[9].combine_first(prm_lg_next_df.loc[22])
prm_lg_next_df.loc[12] = prm_lg_next_df.loc[12].combine_first(prm_lg_next_df.loc[35])
prm_lg_next_df.loc[17] = prm_lg_next_df.loc[17].combine_first(prm_lg_next_df.loc[28])
```

```python
prm_lg_combined_df = pd.concat([prm_lg_combined_df.dropna(), prm_lg_next_df.dropna()])
```

```python
prm_lg_combined_df=prm_lg_combined_df.reset_index(drop=True)
```

```python
champ_league_data_df = pd.read_csv('..\\..\\Consolidated Data\\Champions_League_CLub_Level.csv')

champ_league_data_df.iloc[10, champ_league_data_df.columns.get_loc('Team')] = 'Man Utd'
champ_league_data_df.iloc[10, champ_league_data_df.columns.get_loc('Team')] = 'Tottenhamr'
champ_league_data_df.iloc[10, champ_league_data_df.columns.get_loc('Team')] = 'Manc City'

prm_lg_combined_df=prm_lg_combined_df.merge(champ_league_data_df,how='left',on='Team')

prm_lg_combined_df = prm_lg_combined_df.fillna(0)
```

```python
europa_league_data_df = pd.read_csv('..\\..\\Consolidated Data\\Europa_League_CLub_Level.csv')

europa_league_data_df.iloc[7, champ_league_data_df.columns.get_loc('Team')] = 'Man Utd'

prm_lg_combined_df=prm_lg_combined_df.merge(europa_league_data_df,how='left',on='Team')

prm_lg_combined_df = prm_lg_combined_df.fillna(0)
```

```python
prm_lg_combined_df.columns = ['Team', 'Avg_Pos', 'Total Market Value', 'League', 'Range1_Win_CL',
       'Range2_Win_CL', 'Range3_Win_CL', 'Range4_Win_CL', 'Range5_Win_CL',
       'Range1_Run_CL', 'Range2_Run_CL', 'Range3_Run_CL', 'Range4_Run_CL',
       'Range5_Run_CL', 'Range1_SF_CL', 'Range2_SF_CL', 'Range3_SF_CL',
       'Range4_SF_CL', 'Range5_SF_CL', 'Range1_Win_EL', 'Range2_Win_EL',
       'Range3_Win_EL', 'Range4_Win_EL', 'Range5_Win_EL', 'Range1_Run_EL',
       'Range2_Run_EL', 'Range3_Run_EL', 'Range4_Run_EL', 'Range5_Run_EL',
       'Range1_SF_EL', 'Range2_SF_EL', 'Range3_SF_EL', 'Range4_SF_EL',
       'Range5_SF_EL']
```

```python
world_cup_data_df = pd.read_csv('..\\..\\Consolidated Data\\World Cup Data at Club Level.csv')

club_string = world_cup_data_df['Club']
for i in range(0,len(club_string)):
    club_string[i]=unicodedata.normalize('NFKD', club_string[i]).encode('ASCII', 'ignore').decode('utf-8')

world_cup_data_df['Club'] = club_string

world_cup_data_df.columns = ['Team','League','WC_Player_Count']

world_cup_data_df[world_cup_data_df['League'] ==  'Premier League']
```

```python
prm_lg_temp_df=prm_lg_combined_df.merge(world_cup_data_df,how='outer',on='Team')

for i in range(prm_lg_temp_df[prm_lg_temp_df['Total Market Value'].isnull()].head(1).index[0],prm_lg_temp_df[prm_lg_temp_df['Total Market Value'].isnull()].tail(1).index[0]):
    for j in range(prm_lg_temp_df[prm_lg_temp_df['WC_Player_Count'].isnull()].head(1).index[0],prm_lg_temp_df[prm_lg_temp_df['WC_Player_Count'].isnull()].tail(1).index[0]):
        search = re.search(prm_lg_temp_df.loc[j]['Team'],prm_lg_temp_df.loc[i]['Team'])
        if search is not None:
            clean = search.group()
            prm_lg_temp_df.loc[i] = prm_lg_temp_df.loc[i].combine_first(prm_lg_temp_df.loc[j])
            break
```

```python
prm_lg_temp_df[prm_lg_temp_df['WC_Player_Count'].isnull()][['Team','WC_Player_Count']]
```

```python
prm_lg_temp_df.iloc[8, prm_lg_temp_df.columns.get_loc('WC_Player_Count')] = 8
prm_lg_temp_df.iloc[10, prm_lg_temp_df.columns.get_loc('WC_Player_Count')] = 16
prm_lg_temp_df.iloc[11, prm_lg_temp_df.columns.get_loc('WC_Player_Count')] = 11
prm_lg_temp_df.iloc[16, prm_lg_temp_df.columns.get_loc('WC_Player_Count')] = 12
prm_lg_temp_df.iloc[18, prm_lg_temp_df.columns.get_loc('WC_Player_Count')] = 3
prm_lg_temp_df.iloc[19, prm_lg_temp_df.columns.get_loc('WC_Player_Count')] = 2
```

```python
prm_lg_temp_df=prm_lg_temp_df.drop(['League_y'],axis=1)
prm_lg_temp_df['WC_Player_Count']=prm_lg_temp_df['WC_Player_Count'].fillna(0)
prm_lg_temp_df=prm_lg_temp_df.dropna()
prm_lg_combined_df = prm_lg_temp_df
```

```python
top_50_data_df = pd.read_csv('..\\..\\Consolidated Data\\Top 50 Players Club Level.csv')
top_50_data_df['Value in GBP'] = top_50_data_df['Value in GBP'].map(lambda x: x.lstrip('£').rstrip('m'))
x=top_50_data_df.groupby('Team').agg({'Player':'count'})
x=x.reset_index()
x.iloc[11, x.columns.get_loc('Team')] = 'Man City'
x.iloc[12, x.columns.get_loc('Team')] = 'Man Utd'
x.iloc[16, x.columns.get_loc('Team')] = 'Tottenham'
prm_lg_temp_df=prm_lg_temp_df.merge(x,how='left',on='Team')
prm_lg_combined_df = prm_lg_temp_df.fillna(0)
prm_lg_combined_df=prm_lg_combined_df.drop(prm_lg_combined_df.index[[20,21,22]])
```

```python
fa_cup_df = pd.read_csv('..\\..\\Consolidated Data\\FA Cup Records.csv')
fa_winner = fa_cup_df[['Season','Winner']].groupby('Winner').count()
fa_winner = fa_winner.reset_index()
fa_winner.columns = ['Team','FA_Wins']
fa_winner.iloc[3, x.columns.get_loc('Team')] = 'Man City'
fa_winner.iloc[4, x.columns.get_loc('Team')] = 'Man Utd'
prm_lg_temp_df=prm_lg_combined_df
prm_lg_temp_df= pd.merge(prm_lg_temp_df,fa_winner,how='left',on='Team')
prm_lg_temp_df=prm_lg_temp_df.fillna(0)
```

```python
fa_runner = fa_cup_df[['Season','Runner']].groupby('Runner').count()
fa_runner = fa_runner.reset_index()
fa_runner.columns = ['Team','FA_Runners']
fa_runner.iloc[7, x.columns.get_loc('Team')] = 'Man City'
fa_runner.iloc[8, x.columns.get_loc('Team')] = 'Man Utd'
fa_runner.iloc[12, x.columns.get_loc('Team')] = 'West Ham'
prm_lg_temp_df= pd.merge(prm_lg_temp_df,fa_runner,how='left',on='Team')
prm_lg_temp_df=prm_lg_temp_df.fillna(0)
prm_lg_combined_df=prm_lg_temp_df
```
