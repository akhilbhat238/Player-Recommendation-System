```python
import numpy as np
import pandas as pd
```

```python
bl_07_df = pd.read_csv('..\\..\\Consolidated Data\\English_Premiership_2007.csv')
bl_08_df = pd.read_csv('..\\..\\Consolidated Data\\English_Premiership_2008.csv')
bl_09_df = pd.read_csv('..\\..\\Consolidated Data\\English_Premiership_2009.csv')
bl_10_df = pd.read_csv('..\\..\\Consolidated Data\\English_Premiership_2010.csv')
bl_11_df = pd.read_csv('..\\..\\Consolidated Data\\English_Premiership_2011.csv')
bl_12_df = pd.read_csv('..\\..\\Consolidated Data\\English_Premiership_2012.csv')
bl_13_df = pd.read_csv('..\\..\\Consolidated Data\\English_Premiership_2013.csv')
bl_14_df = pd.read_csv('..\\..\\Consolidated Data\\English_Premiership_2014.csv')
bl_15_df = pd.read_csv('..\\..\\Consolidated Data\\English_Premiership_2015.csv')
bl_16_df = pd.read_csv('..\\..\\Consolidated Data\\English_Premiership_2016.csv')
bl_17_df = pd.read_csv('..\\..\\Consolidated Data\\English_Premiership_2017.csv')
bl_18_df = pd.read_csv('..\\..\\Consolidated Data\\English_Premiership_2018.csv')
bl_19_df = pd.read_csv('..\\..\\Consolidated Data\\English_Premiership_2019.csv')
```

```python
current_team_list = ['Arsenal',
'Aston Villa',
'Bournemouth',
'Brighton',
'Burnley',
'Chelsea',
'Crystal Palace',
'Everton',
'Leicester',
'Liverpool',
'Man City',
'Man Utd',
'Newcastle',
'Norwich',
'Sheffield Utd',
'Southampton',
'Tottenham',
'Watford',
'West Ham',
'Wolverhampton']
```

```python
bl_combined_df = pd.DataFrame(current_team_list, columns =['Team'])
```

```python
bl_combined_df = bl_combined_df.merge(bl_19_df[['Team','Pos']],how='left',on='Team')
```

```python
bl_combined_df = bl_combined_df.merge(bl_18_df[['Team','Pos']],how='left',on='Team')
bl_combined_df = bl_combined_df.merge(bl_17_df[['Team','Pos']],how='left',on='Team')
bl_combined_df = bl_combined_df.merge(bl_16_df[['Team','Pos']],how='left',on='Team')
bl_combined_df = bl_combined_df.merge(bl_15_df[['Team','Pos']],how='left',on='Team')
bl_combined_df = bl_combined_df.merge(bl_14_df[['Team','Pos']],how='left',on='Team')
bl_combined_df = bl_combined_df.merge(bl_13_df[['Team','Pos']],how='left',on='Team')
bl_combined_df = bl_combined_df.merge(bl_12_df[['Team','Pos']],how='left',on='Team')
bl_combined_df = bl_combined_df.merge(bl_11_df[['Team','Pos']],how='left',on='Team')
bl_combined_df = bl_combined_df.merge(bl_10_df[['Team','Pos']],how='left',on='Team')
bl_combined_df = bl_combined_df.merge(bl_09_df[['Team','Pos']],how='left',on='Team')
bl_combined_df = bl_combined_df.merge(bl_08_df[['Team','Pos']],how='left',on='Team')
bl_combined_df = bl_combined_df.merge(bl_07_df[['Team','Pos']],how='left',on='Team')
```

```python
bl_combined_df = bl_combined_df.fillna(21)
bl_combined_df.columns = ['Team','Pos_2019','Pos_2018','Pos_2017','Pos_2016','Pos_2015','Pos_2014','Pos_2013','Pos_2012','Pos_2011','Pos_2010','Pos_2009','Pos_2008','Pos_2007']
```

```python
bl_combined_df['Avg_Pos'] = bl_combined_df.mean(axis=1)
bl_final_df = bl_combined_df[['Team','Avg_Pos']]
bl_final_df.to_csv('..\\..\\Consolidated Data\\Premier League Average Positions.csv',index=False)
```
