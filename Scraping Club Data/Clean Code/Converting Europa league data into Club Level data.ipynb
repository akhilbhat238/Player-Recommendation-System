```python
import numpy as np
import pandas as pd
```

```python
europa_league_df = pd.read_csv('..\\Consolidated Data\\Europa_League_Data.csv')
el_in_scope_df = europa_league_df.loc[europa_league_df.loc[europa_league_df['Season']==2005].index[0]:europa_league_df.loc[europa_league_df['Season']==2019].index[0]]
el_teams = el_in_scope_df['Winners'].append(el_in_scope_df['Runner Up']).append(el_in_scope_df['SF1']).append(el_in_scope_df['SF2'])
el_teams = el_teams.unique()
```

```python
range1_start = int(el_in_scope_df['Season'].max())
range2_start = range1_start-3
range3_start = range2_start-3
range4_start = range3_start-3
range5_start = range4_start-3
range_dict = {"range1_start":range1_start,"range2_start":range2_start,"range3_start":range3_start,"range4_start":range4_start,"range5_start":range5_start}
```

```python
final_df = pd.DataFrame(columns=['Team','Range1_Win','Range2_Win','Range3_Win','Range4_Win','Range5_Win','Range1_Run','Range2_Run','Range3_Run','Range4_Run','Range5_Run','Range1_SF','Range2_SF','Range3_SF','Range4_SF','Range5_SF'])
```

```python
final_df['Team'] = el_teams
final_df = final_df.set_index('Team')
```

```python
final_df=final_df.fillna(0)
```

```python
el_range_1 = el_in_scope_df.loc[45:47]
range1_winners = el_range_1.groupby(['Winners']).count()['Season']
range1_runners = el_range_1.groupby(['Runner Up']).count()['Season']
range1_sf1 = el_range_1.groupby(['SF1']).count()['Season']
range1_sf2 = el_range_1.groupby(['SF2']).count()['Season']
range1_sf = range1_sf1.append(range1_sf2)
range1_sf_df = range1_sf.to_frame()
range1_sf_df.index.name = 'Team'
range1_sf = range1_sf_df.groupby(['Team']).sum()['Season']
```

```python
df = range1_winners.to_frame().T
df.columns
for i in df.columns:
    final_df.loc[i]['Range1_Win'] = range1_winners[i]

df = range1_runners.to_frame().T
df.columns
for i in df.columns:
    final_df.loc[i]['Range1_Run'] = range1_runners[i]

df = range1_sf.to_frame().T
df.columns
for i in df.columns:
    final_df.loc[i]['Range1_SF'] = range1_sf[i]
```
