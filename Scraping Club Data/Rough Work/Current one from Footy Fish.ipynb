```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get("https://thefishy.co.uk/leaguetable.php?table=79&season=16")
soup = BeautifulSoup(page.content,'html.parser')
```

```python
tables = soup.findAll("table")
```

```python
table_to_use = tables[5]
```

```python
table_header = table_to_use.find_all('th')
```

```python
header_list = []
for i in table_header:
    header_list.append(i.contents[0])
```

```python
header_list
```

```python
table_data = table_to_use.find_all('td')
```

```python
table_data_clean = []
for i in table_data:
    table_data_clean.append(i.contents[0])
```

```python
table_data_clean
```

```python
inner_list=[]
for i in range(0,10):
    inner_list.append(table_data_clean[i])
```

```python
inner_list
```

```python
def divide_chunks(l, n): 
    for i in range(0, len(l), n):  
        yield l[i:i + n]
```

```python
n = 10  
x = list(divide_chunks(table_data_clean, n)) 
```

```python
df = pd.DataFrame(x, columns =header_list) 
```

```python
df
```

```python
df_team = df['Team']
```

```python
df_team[0].contents[0]
```

```python
for i in range(0,len(df_team)):
    df_team[i] = df_team[i].contents[0]
```

```python
df['Team'] = df_team
```
