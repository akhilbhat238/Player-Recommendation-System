```python
import numpy as np
import pandas as pd
```

```python
bl_06_df = pd.read_csv('..\\Consolidated Data\\Bundesliga_2006.csv')
bl_07_df = pd.read_csv('..\\Consolidated Data\\Bundesliga_2007.csv')
```

```python
bl_06_df['Team'][2]='Hamburger'
bl_06_df['Team'][3]='Schalke'
bl_06_df['Team'][4]='Leverkusen'
bl_06_df['Team'][5]='Hertha Berlin'
bl_06_df['Team'][7]='Nurnberg'
bl_06_df['Team'][8]='VFB Stuttgart'
bl_06_df['Team'][9]='Borussia Monchengladbach'
bl_06_df['Team'][10]='Mainz'
bl_06_df['Team'][14]='Wolfsburg'
```

```python
for i in range(0,len(bl_06_df['Team'])):
    bl_06_df['Team'][i] = bl_06_df['Team'][i].replace(' (C)','')
    bl_06_df['Team'][i] = bl_06_df['Team'][i].replace(' (R)','')
```

```python
for i in range(0,len(bl_07_df['Team'])):
    bl_07_df['Team'][i] = bl_07_df['Team'][i].replace(' (C)','')
    bl_07_df['Team'][i] = bl_07_df['Team'][i].replace(' (R)','')
```

```python
bl_07_df
```

```python
bl_combined_df = bl_06_df[['Team','Pos']].merge(bl_07_df[['Team','Pos']],how='outer',on='Team')
```

```python
bl_combined_df = bl_combined_df.rename(columns={"Pos_x": "Pos_2006", "Pos_y": "Pos_2007"})
```

```python
bl_combined_df
```
