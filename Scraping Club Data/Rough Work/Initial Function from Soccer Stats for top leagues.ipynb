```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
def build_data_frame(league,season):
    league_url = 'https://www.soccerstats.com/latest.asp?league='+league+'_'+str(season)
    page = requests.get(league_url)
    soup = BeautifulSoup(page.content,'html.parser')
    table_element = soup.find('table',id='btable')
    table_rows = table_element.find_all('tr')
    header_row = table_rows[0].find_all('th')
    column_names = []
    for ele in header_row:
        column_names.append(ele.text.strip().upper())
    column_names[0] = 'RANKING'
    column_names[1] = 'TEAM'
    column_names.remove('FORM')
    column_names.remove('PPG')
    column_names.remove('CS')
    column_names.remove('FTS')
    if season > 2017:
        column_names.remove('LAST 8')
    df = pd.DataFrame(columns=column_names)
    x=1
    for i in range(1,len(table_rows)):
        team_row = table_rows[x].find_all('td')
        team_col_val = []
        for ele in team_row:
            team_col_val.append(ele.text.strip())
        del team_col_val[10:]
        df.loc[i] = team_col_val
        x=x+7
        if (league == 'netherlands' or league == 'germany'):
            if i==18:
                break;
        else:
            if i==20:
                break
    dataframe = df.set_index('RANKING')
    return dataframe
```

```python
df_spain_2015 = build_data_frame('spain',2015)
```
