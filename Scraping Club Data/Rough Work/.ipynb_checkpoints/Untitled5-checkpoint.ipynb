```python
bl_06_df = pd.read_csv('..\\Consolidated Data\\Bundesliga_2006.csv')
bl_07_df = pd.read_csv('..\\Consolidated Data\\Bundesliga_2007.csv')
bl_08_df = pd.read_csv('..\\Consolidated Data\\Bundesliga_2008.csv')
bl_09_df = pd.read_csv('..\\Consolidated Data\\Bundesliga_2009.csv')
bl_10_df = pd.read_csv('..\\Consolidated Data\\Bundesliga_2010.csv')
bl_11_df = pd.read_csv('..\\Consolidated Data\\Bundesliga_2011.csv')
bl_12_df = pd.read_csv('..\\Consolidated Data\\Bundesliga_2012.csv')
bl_13_df = pd.read_csv('..\\Consolidated Data\\Bundesliga_2013.csv')
bl_14_df = pd.read_csv('..\\Consolidated Data\\Bundesliga_2014.csv')
bl_15_df = pd.read_csv('..\\Consolidated Data\\Bundesliga_2015.csv')
bl_16_df = pd.read_csv('..\\Consolidated Data\\Bundesliga_2016.csv')
bl_17_df = pd.read_csv('..\\Consolidated Data\\Bundesliga_2017.csv')
bl_18_df = pd.read_csv('..\\Consolidated Data\\Bundesliga_2018.csv')
bl_19_df = pd.read_csv('..\\Consolidated Data\\Bundesliga_2019.csv')
```

```python
bl_06_df['Team'][2]='Hamburger'
bl_06_df['Team'][3]='Schalke'
bl_06_df['Team'][4]='Leverkusen'
bl_06_df['Team'][5]='Hertha Berlin'
bl_06_df['Team'][7]='Nurnberg'
bl_06_df['Team'][8]='VFB Stuttgart'
bl_06_df['Team'][9]='Borussia Monchengladbach'
bl_06_df['Team'][10]='Mainz'
bl_06_df['Team'][14]='Wolfsburg'
```

```python
current_team_list = ['Augsburg',
'Bayern Munich',
'Borussia Dortmund',
'Borussia Monchengladbach',
'Eintracht Frankfurt',
'FC Koln',
'Fortuna Dusseldorf',
'Freiburg',
'Hertha Berlin',
'Hoffenheim',
'Leverkusen',
'Mainz',
'RB Leipzig',
'SC Paderborn 07',
'Schalke',
'Union Berlin',
'Werder Bremen',
'Wolfsburg']
```

```python
bl_combined_df = pd.DataFrame(current_team_list, columns =['Team'])
```

```python
bl_combined_df = bl_combined_df.merge(bl_19_df[['Team','Pos']],how='left',on='Team')
```

```python
bl_combined_df = bl_combined_df.merge(bl_18_df[['Team','Pos']],how='left',on='Team')
bl_combined_df = bl_combined_df.merge(bl_17_df[['Team','Pos']],how='left',on='Team')
bl_combined_df = bl_combined_df.merge(bl_16_df[['Team','Pos']],how='left',on='Team')
bl_combined_df = bl_combined_df.merge(bl_15_df[['Team','Pos']],how='left',on='Team')
bl_combined_df = bl_combined_df.merge(bl_14_df[['Team','Pos']],how='left',on='Team')
bl_combined_df = bl_combined_df.merge(bl_13_df[['Team','Pos']],how='left',on='Team')
bl_combined_df = bl_combined_df.merge(bl_12_df[['Team','Pos']],how='left',on='Team')
bl_combined_df = bl_combined_df.merge(bl_11_df[['Team','Pos']],how='left',on='Team')
bl_combined_df = bl_combined_df.merge(bl_10_df[['Team','Pos']],how='left',on='Team')
bl_combined_df = bl_combined_df.merge(bl_09_df[['Team','Pos']],how='left',on='Team')
bl_combined_df = bl_combined_df.merge(bl_08_df[['Team','Pos']],how='left',on='Team')
bl_combined_df = bl_combined_df.merge(bl_07_df[['Team','Pos']],how='left',on='Team')
bl_combined_df = bl_combined_df.merge(bl_06_df[['Team','Pos']],how='left',on='Team')
```

```python
bl_combined_df = bl_combined_df.fillna(0)
```

```python
bl_combined_df.columns = ['Team','Pos_2019','Pos_2018','Pos_2017','Pos_2016','Pos_2015','Pos_2014','Pos_2013','Pos_2012','Pos_2011','Pos_2010','Pos_2009','Pos_2008','Pos_2007','Pos_2006']
```

```python
bl_combined_df
```

```python
bl_combined_df['Avg_Pos'] = bl_combined_df.mean(axis=1)
```

```python
bl_final_df = bl_combined_df[['Team','Avg_Pos']]
```

```python
bl_final_df.to_csv('..\\Consolidated Data\\Bundesliga Average Positions.csv',index=False)
```
